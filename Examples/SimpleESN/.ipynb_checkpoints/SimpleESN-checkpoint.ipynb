# [Simple ESN from  Mantas Lukoševičius ](https://mantas.info/code/simple_esn/)
- Apply RC on Mackey-Glass Delay Differential Equations

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge,LinearRegression
from scipy import linalg # for eigenvalue
np.random.seed(42)
```

## Import Data

```python
data = np.loadtxt('./mgdata.txt')[:,1] #1201 data points
# data = np.loadtxt('./MackeyGlass_t17.txt') # 10000 data points
plt.figure(figsize=(20,5),dpi=200)
plt.plot(data[:1000])
plt.title("Mackey-Glass delay differential equations solution with delay=17")
#1-D time series
inSize = outSize = 1
trainval_split=0.7
```

## Generate the reservoir

```python
resSize = 1000
alpha = 0.3 # leaking rate

#Creating input-reservoir and reservoir-reservoir connections

Win = np.random.rand(resSize,1+inSize) - 0.5
W = np.random.rand(resSize,resSize) - 0.5 

# normalizing and setting spectral radius (correct, slow):
print('Computing spectral radius...')
rhoW = abs(linalg.eig(W)[0]).max()
print(f'Done: {rhoW}')

print('Scaling matrix with spectral radius...')
W *= 1.25 / rhoW

print('Recalculating spectral radius...')
rhoW = abs(linalg.eig(W)[0]).max()
print(f'Done: {rhoW}')
```

### Leaking rate $\alpha$

$\alpha$ can be regarded as the time interval in the continuous world between two consecutive time steps in the discrete realization. Small alpha $\rightarrow$ slow dynamics of x(n) $\rightarrow$ longer short term memory.

### Recurrent reservoir unit $\textbf x(n)$ update rule
$\tilde{\textbf x}(n) = \tanh (\textbf{W}^{in} \cdot [1;u(n)] + \textbf{W} \cdot \textbf x(n − 1)) \\
\textbf x(n) = (1 − α)\textbf x(n − 1) + \alpha \tilde{\textbf x}(n)$


```python
# Here X is the reservoir output, which the readout weights will be trained on vias ridge regression
trainLen = int(trainval_split * data.size)
valLen = data.size - trainLen
initLen = 100 #We initialize the reservoir states for 100 timesteps

#Array shapes: (column vector, time)

y_train = data[None,initLen+1:trainLen+1]
X = np.zeros((1+inSize+resSize,trainLen-initLen)) 
x = np.zeros((resSize,1))  #reservoir states
for t in range(trainLen):
    u = data[t]
    x = (1-alpha)*x + alpha*np.tanh( np.dot( Win, np.vstack((1,u)) ) + np.dot( W, x ) )
    if t >= initLen:
        X[:,t-initLen] = np.vstack((1,u,x)).ravel()
```

## Ridge Regression

```python
regr = Ridge(1e-8, tol=0.1, fit_intercept=False,solver="auto")
regr.fit(X.T,y_train.T)
Wout = regr.coef_
print(f"MSE train: {((np.dot(Wout,X)-y_train)**2).mean()}")
```

## Validation

```python
#Generative Mode
#--------------
# we use values of x from the training continuing from where we left off.
# predicted outputs of the network become input for the next time step.

X_val = np.zeros((resSize,valLen)) #Just for plotting the reservoir states later on
y_val = np.zeros((outSize,valLen))
u = data[trainLen]
for t in range(valLen):
    x = (1-alpha)*x + alpha*np.tanh( np.dot( Win, np.vstack((1,u)) ) + np.dot( W, x ) )
    X_val[:,t] = x.ravel()
    y = np.dot( Wout, np.vstack((1,u,x)) )
    y_val[:,t] = y
    # generative mode:
    u = y
    ## this would be a predictive mode:
    #u = data[trainLen+t+1] 

errorLen = valLen
mse = ((data[trainLen+1:trainLen+errorLen+1] - y_val[:,1:errorLen].ravel())**2).mean()
print('MSE validation: ' + str( mse ))
```

### Plots

#### Prediction

```python
plt.figure(figsize=(20,5),dpi=200)
plt.plot(data[trainLen:trainLen+errorLen],'b',label='Ground Truth')
plt.plot(y_val[:errorLen].ravel(),'r.',label='Prediction');
plt.legend(fontsize=15);
```
